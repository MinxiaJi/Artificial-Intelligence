```python
# CSGA 6525 Artificial Intelligence 
# Fall 2017
# Assignment #2
# Minxia Ji
# Local beam search applied on 8-queen problem 
import numpy as np
###############
# question (a)#
###############
#write the code to determine how many pairs of attacking queens there are
def attacking_queens(states):
    count = 0
    for i in range(len(states)):
        current = states[i]
        following = states[i+1:]
        for j,state in enumerate(following):
        # count possible attacking queens that on the same row
            if current == state:
                count += 1
        # count possible attacking queens that on the diagnol
            if abs(int(current)-int(state))==j+1:
                count += 1
    return count

###############
# question (b)#
###############
def initialize_state(k):
    initial_states = []
    for i in range(k):
        # generate 8 queens position on rows
        temp = np.random.randint(0,8,8)
        state = ''
        for n in temp:
            state += str(n)
        initial_states.append(state)
    return initial_states

# based on current states generate possible next states
def generate_new_states(current_state):
    new_states = []
    for i,row in enumerate(current_state):
        positions = [position for position in range(8)]
        # pop current row where the queen currently located on 
        positions.pop(int(row))
        temp = []
        for j in positions:
            # new state equals to a string replaced the one substring with the new substring
            new_state = current_state[:i]+str(j)+current_state[i+1:]
            temp.append(new_state)
        new_states.append(temp)
    return new_states



#  Local beam search with a termination condition:either successful (cost=0) or not (local cost minimum)
def local_beam_search(k,PRINT=True):
    # current K states
    current_states = initialize_state(k)
    # 'position':how many attacking queens pair dictionary
    state_aqpairs = {}
    # select K new states
    new_states = []
    continue_iter = True
    min_before = 8
    solutions = []
    while continue_iter:
        
        for i in current_states:
            # compute how many attacking queens the each current state contains:
            state_aqpairs[i]=attacking_queens(i)
            # generate 56*k states
            new_states.append(generate_new_states(i))
       # calculate attacking queen pairs on these states
        for j in new_states:
            for moves in j:
                for move in moves:
                    state_aqpairs[move] = attacking_queens(move)
        min_current = min(state_aqpairs.values())
        
        # if we find goal state:0
        if min_current == 0:
            continue_iter = False
            for position,aq in state_aqpairs.items():
                if aq == 0:
                    solutions.append(position)
            return solutions
        if min_current >= min_before:
            continue_iter = False
            return min_before
        min_before = min_current

        current_states = sorted(state_aqpairs, key= state_aqpairs.get)[:k]
        
###############
# question (c)#
###############
# question (c): Show the performance of each value of k by generating a large number of 8 queen problems
# k1: beam size 1
# k2: beam size 2
# k3: beam size 3
# trials: no. of trials you want to conduct when examing the performance of different beam size
# user_input:For solution of running one time LBS search please input 'S'; 
# ###########For performance please input 'P'.
def measure_performance(k1,k2,k3,trials,user_input):
    iter_time = 0
    minima_k1 = 0
    minima_k2 = 0
    minima_k3 = 0
    solutions_k1 = 0
    solutions_k2 = 0
    solutions_k3 = 0
    if user_input == 'S':
        result_k1 = local_beam_search(k1)
        result_k2 = local_beam_search(k2)
        result_k3 = local_beam_search(k3)
        if type(result_k1) is int:
            print 'Local beam search with beam size {} found local minimum: {}'.format(k1,result_k1)
        else:
            print 'The solution is',result_k1
        if type(result_k2) is int:
            print 'Local beam search with beam size {} found local minimum: {}'.format(k2,result_k2)
        else:
            print 'The solution is',result_k2
        if type(result_k3) is int:
            print 'Local beam search with beam size {} found local minima: {}'.format(k3,result_k3)
        else:
            print 'The solution is',result_k3
    if user_input == 'P':
        while iter_time < trials:
            result_k1 = local_beam_search(k1)
            if type(result_k1) is int:
                minima_k1 += result_k1
            else:
                solutions_k1 += len(result_k1)
            result_k2 = local_beam_search(k2)
            if type(result_k2) is int:
                minima_k2 += result_k2
            else:
                solutions_k2 += len(result_k2)
            result_k3 = local_beam_search(k3)
            if type(result_k3) is int:
                minima_k3 += result_k3
            else:
                solutions_k3 += len(result_k3)

                        
        
            iter_time += 1
        print 'Run {} times of Loal beam search when beam size = {} stucked in average local minimum :{}'.format(trials,k1,float(minima_k1)/float(trials))
        print 'Run {} times of Loal beam search when beam size = {} stucked in average local minimum :{}'.format(trials,k2,float(minima_k2)/float(trials))
        print 'Run {} times of Loal beam search when beam size = {} stucked in average local minimum :{}'.format(trials,k3,float(minima_k3)/float(trials))
        print 'Run {} times of Loal beam search when beam size = {} found {} solutions'.format(trials,k1,solutions_k1)
        print 'Run {} times of Loal beam search when beam size = {} found {} solutions'.format(trials,k2,solutions_k2)
        print 'Run {} times of Loal beam search when beam size = {} found {} solutions'.format(trials,k3,solutions_k3)
```
